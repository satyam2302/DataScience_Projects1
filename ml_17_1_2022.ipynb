#Python Basics


```python
listOfNumbers = [1, 2, 3, 4, 5, 6]

for number in listOfNumbers:
  if(number % 2 == 0):
    print(number, " is even")
  else:
    print(number,"is odd")

print("all done")
```

#Functions

```python
def SquareIt(x):
  return x*x

print(SquareIt(2))
```

```python
def DoSomething(f,x):
  return f(x)

print(DoSomething(SquareIt,3))  
```

```python
#lambda function let you inline simple function
print(DoSomething(lambda x: x*x*x,4))
```

```python
print(SquareIt(lambda x: x*x*x))
print(SquareIt(2))
```
